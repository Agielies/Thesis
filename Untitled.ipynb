```python
import numpy as np


#Symmettric payoffs

# experimentail appproach
def sharpe_precision_exp(precision):
    """
    Calculates the sharpe ratio for a given precision in a simulation
    """
    out = []
    p = precision
    
    for i in range(100000):
        rnd = np.random.binomial(n = 1,p = p)
        x = (1 if rnd == 1 else -1)
        out.append(x)

    return np.mean(out)/np.std(out)



def sharpe_precision_num_trades(precision,num_trades):
    """
    Analytical solution for the sharpe ratio given the precisoin and number of trades.
    """
    p = precision
    
    sharpe = ((2*p -1) / (2*np.sqrt(p*(1-p))))*np.sqrt(num_trades)

    return sharpe

sharpe_precision_num_trades(0.55,396)
                
```

```python
# lets try now when we have assymetric pay offs
pi_pos = 0.1
pi_neg =-0.1

def sharpe_precision_num_trades_assym(precision,num_trades,pi_pos, pi_neg):
    """
    Analytical solution for the sharpe ratio given the precisoin and number of trades.
    """
    p = precision
    
    sharpe = ((p*(pi_pos-pi_neg) +pi_neg) / ((pi_pos-pi_neg)*np.sqrt(p*(1-p))))*np.sqrt(num_trades)
    
    

    return sharpe

sharpe_precision_num_trades_assym(0.55,396,0.01,-0.01)
```

```python
import matplotlib.pyplot as plt



# lets plot the sharpe ratio as a function of the ration of expected payoffs to expected losses
x_values = 10

# set the values for the pi_neg
pi_neg = -0.01

sharpes = [0.0]*x_values
pi_pos_list =[0.0]*x_values
ratio = [0.0]*x_values
# find the values of the sharpe ratio and the the ration of payoffs
fig, ax = plt.subplots()

for j in range(50,75,5):
    prec = j/100
    for i in range(1,x_values+1,1):
        a = i/400
        pi_pos_list[i-1] = a
        sharpes[i-1] = sharpe_precision_num_trades_assym(prec
                                                         ,200,pi_pos_list[i-1],pi_neg)
        
        ratio[i-1] = -(pi_pos_list[i-1]/pi_neg)**(1)

    y = sharpes
    x = ratio
    ax.plot(x,y, label = 'Precision {}'.format(j/100),linewidth=2)



#ax.plot(x, y, linewidth=2.0)
plt.ylabel('Sharpe ratio')
plt.xlabel('Ratio of exepected postive payouts to negative payouts')
plt.legend()
plt.show()  


# lets plot the sharpe ratio as a function of the ration of expected payoffs to expected losses
x_values = 10

# set the values for the pi_neg
pi_neg = -0.01

sharpes = [0.0]*x_values
pi_pos_list =[0.0]*x_values
ratio = [0.0]*x_values
# find the values of the sharpe ratio and the the ration of payoffs
fig, ax = plt.subplots()

for j in range(50,75,5):
    prec = j/100
    for i in range(1,x_values+1,1):
        a = i/400
        pi_pos_list[i-1] = a
        sharpes[i-1] = sharpe_precision_num_trades_assym(prec
                                                         ,200,pi_pos_list[i-1],pi_neg)
        
        ratio[i-1] = pi_pos_list[i-1] + pi_neg

    y = sharpes
    x = ratio
    ax.plot(x,y, label = 'Precision {}'.format(j/100),linewidth=2)



#ax.plot(x, y, linewidth=2.0)
plt.ylabel('Sharpe ratio')
plt.xlabel('Ratio of exepected postive payouts to negative payouts')
plt.legend()
plt.show()  
```

```python
# 
def precision(n,theta,pi_pos,pi_neg):
    a = (n+theta**2)*((pi_pos - pi_neg)**2)
    b = (2*n*pi_neg- (theta**2)*(pi_pos - pi_neg))*(pi_pos - pi_neg)
    c = n* pi_neg**2
    p = (-b + np.sqrt(b**2-4*a*c))/(2*a)

    return p

# lets plot the sharpe ratio as a function of the ration of expected payoffs to expected losses
x_values = 10

# set the values for the pi_neg
pi_neg = -0.01

precisions = [0.0]*x_values
pi_pos_list =[0.0]*x_values
ratio = [0.0]*x_values

fig, ax = plt.subplots()

for j in range(1,5):
    sharpe = j
    for i in range(1,x_values+1,1):
        d = float(i)/100
        pi_pos_list[i-1] = d
        precisions[i-1] = precision(100,sharpe,pi_pos_list[i-1],pi_neg)
        ratio[i-1] = -(pi_pos_list[i-1]/pi_neg)**(-1)


    y = precisions
    x = ratio
    ax.plot(x, y, label = 'Sharpe: {}'.format(j/100),linewidth=2.0)

# plot



plt.ylabel('Precision')
plt.xlabel('Ratio of exepected postive payouts to negative payouts')
   
plt.show()
```
